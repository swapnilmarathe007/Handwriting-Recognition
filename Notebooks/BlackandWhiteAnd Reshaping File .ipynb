# From RGB to Black and White 


```python
# Import all necessary packages
import cv2 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
image_location = "/home/syedjafer/Documents/Handwriting_recognition_svm/images/a/a_00.png"
img = cv2.imread(image_location,0) 
(thresh, im_bw) = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
cv2.imwrite('bw_image.png', im_bw)
```

```python
dim = (100 , 100)
resized = cv2.resize(im_bw, dim, interpolation = cv2.INTER_AREA)
```

```python
plt.imshow(resized)
```

```python
cv2.imwrite('bw_image.png', resized)
```

```python
im = cv2.imread('../testing.png')
im3 = im.copy()

gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(5,5),0)
thresh = cv2.adaptiveThreshold(blur,255,1,1,11,2)


```

```python
count = (cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE))
```

```python
for cnt in count:
    if cv2.contourArea(cnt)>50:
        [x,y,w,h] = cv2.boundingRect(cnt)
        if  h>28:
            cv2.rectangle(im,(x,y),(x+w,y+h),(0,255,0),2)
            roi = thresh[y:y+h,x:x+w]
            roismall = cv2.resize(roi,(10,10))
            roismall = roismall.reshape((1,100))
            roismall = np.float32(roismall)
            plt.imshow(roismall)
```
